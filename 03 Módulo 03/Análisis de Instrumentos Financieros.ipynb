```python
# Importar librerías
import numpy as np
import pandas as pd
from pandas_datareader import data
```

```python
# Descargar la información de: GOOG, VOO, ^GSPC
GOOG=data.DataReader('GOOG', data_source='yahoo',start='2010-1-1')
VOO=data.DataReader('VOO', data_source='yahoo',start='2010-1-1')
SP=data.DataReader('^GSPC', data_source='yahoo',start='2010-1-1')
```

### 2.- Definimos nuestra base final

```python
#Unimos las bases creando una nueva con la fecha
portafolio=pd.merge(GOOG,VOO, on='Date', how='left')
portafolio
```

```python
# Reordenamos las columnas que nos interesan: Adj Close porque es el precio de cierre ajustado el que nos interesa
portafolio=portafolio[['Adj Close_x','Adj Close_y']]
portafolio['GOOG']=portafolio['Adj Close_x']
portafolio['VOO']=portafolio['Adj Close_y']
```

```python
#Eliminamos aquellas que no vamos a utilizar
portafolio.drop('Adj Close_x',axis=1, inplace=True)
portafolio.drop('Adj Close_y',axis=1, inplace=True)
```

```python
portafolio
```

```python
# Generamos los mismos cálculos con SP
portafolio=pd.merge(portafolio,SP, on='Date', how='left')
portafolio
```

```python
# Filtramos las columnas finales
portafolio=portafolio[['GOOG','VOO','Adj Close']]
portafolio['SP']=portafolio['Adj Close']
```

```python
portafolio.columns
```

```python
# Eliminamos la columna que ya renombramos anteriormente
portafolio.drop('Adj Close',axis=1, inplace=True)
```

```python
# Tenemos nuestra base final
portafolio
```

```python
# Guardamos nuestra base final con otro nombre y revisamos el tipo de datos
df=portafolio
```

### 4.- Análisis: rendimientos y volatilidad

#### Ya que hemos limpiado la base, queremos estudiar el rendimiento y su variabilidad por lo que estudiaremos los rendimientos en lugar de los precios puesto que es lo que a nosotros nos interesa al momento de realizar una inversión

```python
# Podemos ver que aquí solo tenemos precios y no rendimientos
df.head()
```

```python
# Calculamos los rendimientos
df['Rend_GOOG']=df['GOOG'].pct_change(1)
df['Rend_VOO']=df['VOO'].pct_change(1)
df['Rend_SP']=df['SP'].pct_change(1)
```

```python
# Borramos los precios
df.drop('GOOG',axis=1, inplace=True)
df.drop('VOO',axis=1, inplace=True)
df.drop('SP',axis=1, inplace=True)
```

```python
# Multiplicamos los resultados por 100 puesto que son %
df=df.round(2).mul(100)
df
```

#### Análisis descriptivo estadístico

```python
df.describe(include='all').round(2)
```

```python
#0.10% = 0.001
# Entre más rendimiento mayor riesgo/ volatidalidad (std)
```

#### Un coeficiente que nos indica la relación riesgo retorno es el Coeficiente de variación: te indica qué tan variable es tu instrumento.
#### Calcularemos Coef de Var: STD/Mean

```python
# Calculamos el coeficiente de variación de GOOG
Rend_GOOG_mean=df['Rend_GOOG'].mean()
Rend_GOOG_desvest=df['Rend_GOOG'].std()
Rend_GOOG_coefvar=Rend_GOOG_desvest/Rend_GOOG_mean
print(Rend_GOOG_mean)
print(Rend_GOOG_desvest)
print(Rend_GOOG_coefvar)
```

```python
#Calculamos el coeficiente de variación de VOO
Rend_VOO_mean=df['Rend_VOO'].mean()
Rend_VOO_desvest=df['Rend_VOO'].std()
Rend_VOO_coefvar=Rend_VOO_desvest/Rend_VOO_mean
print(Rend_VOO_mean)
print(Rend_VOO_desvest)
print(Rend_VOO_coefvar)
```

```python
#Calculamos el coeficiente de variación de SP
Rend_SP_mean=df['Rend_SP'].mean()
Rend_SP_desvest=df['Rend_SP'].std()
Rend_SP_coefvar=Rend_SP_desvest/Rend_SP_mean
print(Rend_SP_mean)
print(Rend_SP_desvest)
print(Rend_SP_coefvar)
```

```python
# Calculamos tabla resumen
resumen=pd.DataFrame(data=[[Rend_GOOG_mean,Rend_VOO_mean,Rend_SP_mean],[Rend_GOOG_desvest,Rend_VOO_desvest,Rend_SP_desvest],[Rend_GOOG_coefvar,Rend_VOO_coefvar,Rend_SP_coefvar]],columns=['GOOG','VOO','SP'])
resumen
```

```python
# Cambiamos el nombre del index
resumen.set_index([['Rendimiento','Desvest','Coef de Var']],inplace=True)
resumen
```

### GOOG es la acción con menor variabilidad y con mayor rendimiento, sin embargo, con mayor riesgo.
### VOO tiene mejor rendimeinto que SP y mismo riesgo y menor variablidad por lo que la decisión está entre GOOG y VOO.
### Esta decisión depende de qué tan abierto al riesgo eres
